# Face Detection

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import cv2

alanturing = cv2.imread('Week 4/Practical 4a/AlanTuring.jpg',0)
plt.imshow(alanturing,cmap='gray')
```

```python
gracehopper = cv2.imread('Week 4/Practical 4a/GraceHopper.jpg',0)
plt.imshow(gracehopper,cmap='gray')
```

```python
codebreakers = cv2.imread('Week 4/Practical 4a/CodeBreakers.jpg',0)
plt.imshow(codebreakers,cmap='gray')
```

**Haar Cascade**

```python
face_cascade = cv2.CascadeClassifier('Week 4/Practical 4a/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    
  
    face_img = img.copy()
  
    face_rects = face_cascade.detectMultiScale(face_img) 
    
    for (x,y,w,h) in face_rects: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 10) 
        
    return face_img
    
```

```python
plt.imshow(detect_face(gracehopper),cmap='gray')
```

```python
plt.imshow(detect_face(codebreakers),cmap='gray')
```

```python
plt.imshow(detect_face(alanturing),cmap='gray')
```

**Eye Detection**

```python
#Reset to Frontal Face
face_cascade = cv2.CascadeClassifier('Week 4/Practical 4a/haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('Week 4/Practical 4a/haarcascades/haarcascade_eye_tree_eyeglasses.xml')
```

```python
def detect_eyes(img):
    
    face_img = img.copy()
  
    eyes = eye_cascade.detectMultiScale(face_img) 
    
    
    for (x,y,w,h) in eyes: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 10) 
        
    return face_img
    
```

```python
plt.imshow(detect_eyes(gracehopper),cmap='gray')
```

**Video**

```python
#It is also possible to complete face detection in Video but not on Colab. You will need to take this to Jupyter Notebook running locally.

cap = cv2.VideoCapture(0) 

while True: 
    
   ret, frame = cap.read(0) 
     
   frame = detect_face(frame)
 
   cv2.imshow('Video Face Detection', frame) 
 
   c = cv2.waitKey(1) 
   if c == 27: 
       break 
        
cap.release() 
cv2.destroyAllWindows()
```
